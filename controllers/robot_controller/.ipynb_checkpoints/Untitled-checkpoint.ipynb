```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
 # Initialize the control system
Error = ctrl.Antecedent(np.arange(-100, 100.1, 0.1), 'Error')
DerivativeError = ctrl.Antecedent(np.arange(-50, 50.1, 0.1), 'DerivativeError')
DeltaSpeed = ctrl.Consequent(np.arange(-2.0, 2.01, 0.01), 'DeltaSpeed')
```

```python
Error.automf(3, names=['negative', 'zero', 'positive'])
DerivativeError.automf(3, names=['negative', 'zero', 'positive'])
DeltaSpeed.automf(3, names=['negative', 'zero', 'positive'])
```

```python
Error['negative'] = fuzz.gaussmf(Error.universe, -100, 0)
Error['zero'] = fuzz.gaussmf(Error.universe, 0, 10)
Error['positive'] = fuzz.gaussmf(Error.universe, 100, 15)
```

```python
Error.view()
```

```python
# Error membership functions
Error['VN'] = fuzz.gaussmf(Error.universe, -100, 15)
Error['NB'] = fuzz.trapmf(Error.universe, [-100, -100, -70, -50])
Error['NM'] = fuzz.trimf(Error.universe, [-70, -50, -25])
Error['ZE'] = fuzz.gaussmf(Error.universe, 0, 10)  # Sempitkan range
Error['PM'] = fuzz.trimf(Error.universe, [25, 50, 70])
Error['PB'] = fuzz.trapmf(Error.universe, [50, 70, 100, 100])
Error['VP'] = fuzz.gaussmf(Error.universe, 100, 15)

# Derivative Error membership functions
DerivativeError['VN'] = fuzz.gaussmf(DerivativeError.universe, -50, 7)
DerivativeError['NB'] = fuzz.trapmf(DerivativeError.universe, [-50, -50, -35, -20])
DerivativeError['NM'] = fuzz.trimf(DerivativeError.universe, [-35, -20, -10])
DerivativeError['ZE'] = fuzz.gaussmf(DerivativeError.universe, 0, 5)  # Sempitkan range
DerivativeError['PM'] = fuzz.trimf(DerivativeError.universe, [10, 20, 35])
DerivativeError['PB'] = fuzz.trapmf(DerivativeError.universe, [20, 35, 50, 50])
DerivativeError['VP'] = fuzz.gaussmf(DerivativeError.universe, 50, 7)

# Delta Speed membership functions
DeltaSpeed['KB'] = fuzz.trapmf(DeltaSpeed.universe, [-3, -3, -2.5, -2])
DeltaSpeed['KS'] = fuzz.trimf(DeltaSpeed.universe, [-1.8, -1.2, -0.6])
DeltaSpeed['ST'] = fuzz.trimf(DeltaSpeed.universe, [-0.5, 0, 0.5])
DeltaSpeed['TS'] = fuzz.trimf(DeltaSpeed.universe, [0.6, 1.2, 1.8])
DeltaSpeed['TB'] = fuzz.trapmf(DeltaSpeed.universe, [2, 2.5, 3, 3])
```

```python
Error.view()
DerivativeError.view()
DeltaSpeed.view()
```

```python
rules = [
    ctrl.Rule(Error['NB'] & DerivativeError['NB'], DeltaSpeed['KB']),
    ctrl.Rule(Error['NB'] & DerivativeError['NM'], DeltaSpeed['KB']),
    ctrl.Rule(Error['NB'] & DerivativeError['ZE'], DeltaSpeed['KS']),
    ctrl.Rule(Error['NB'] & DerivativeError['PM'], DeltaSpeed['KS']),
    ctrl.Rule(Error['NB'] & DerivativeError['PB'], DeltaSpeed['ST']),

    ctrl.Rule(Error['NM'] & DerivativeError['NB'], DeltaSpeed['KB']),
    ctrl.Rule(Error['NM'] & DerivativeError['NM'], DeltaSpeed['KS']),
    ctrl.Rule(Error['NM'] & DerivativeError['ZE'], DeltaSpeed['KS']),
    ctrl.Rule(Error['NM'] & DerivativeError['PM'], DeltaSpeed['ST']),
    ctrl.Rule(Error['NM'] & DerivativeError['PB'], DeltaSpeed['ST']),

    ctrl.Rule(Error['ZE'] & DerivativeError['NB'], DeltaSpeed['KS']),
    ctrl.Rule(Error['ZE'] & DerivativeError['NM'], DeltaSpeed['ST']),
    ctrl.Rule(Error['ZE'] & DerivativeError['ZE'], DeltaSpeed['ST']),
    ctrl.Rule(Error['ZE'] & DerivativeError['PM'], DeltaSpeed['TS']),
    ctrl.Rule(Error['ZE'] & DerivativeError['PB'], DeltaSpeed['TS']),

    ctrl.Rule(Error['PM'] & DerivativeError['NB'], DeltaSpeed['ST']),
    ctrl.Rule(Error['PM'] & DerivativeError['NM'], DeltaSpeed['ST']),
    ctrl.Rule(Error['PM'] & DerivativeError['ZE'], DeltaSpeed['TS']),
    ctrl.Rule(Error['PM'] & DerivativeError['PM'], DeltaSpeed['TS']),
    ctrl.Rule(Error['PM'] & DerivativeError['PB'], DeltaSpeed['TB']),

    ctrl.Rule(Error['PB'] & DerivativeError['NB'], DeltaSpeed['ST']),
    ctrl.Rule(Error['PB'] & DerivativeError['NM'], DeltaSpeed['TS']),
    ctrl.Rule(Error['PB'] & DerivativeError['ZE'], DeltaSpeed['TB']),
    ctrl.Rule(Error['PB'] & DerivativeError['PM'], DeltaSpeed['TB']),
    ctrl.Rule(Error['PB'] & DerivativeError['PB'], DeltaSpeed['TB'])
]
```

```python
rules
```

```python
DeltaSpeedControl = ctrl.ControlSystem(rules)
DeltaSpeedSim = ctrl.ControlSystemSimulation(DeltaSpeedControl)
```

```python
# Test dengan nilai masukan
DeltaSpeedSim.input['Error'] = 0.5
DeltaSpeedSim.input['DerivativeError'] = -0.5
DeltaSpeedSim.compute()
```

```python
print(f"Output Delta Speed: {DeltaSpeedSim.output['DeltaSpeed']}")
DeltaSpeed.view(sim=DeltaSpeedSim)
```
